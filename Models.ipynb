```python
def RanForest(Data):
    clf = RandomForestClassifier(n_estimators=500, max_depth=7, random_state=0, min_samples_leaf=1, 
                                 criterion='gini')
    clf.fit(Data[0], Data[2])
    # clf.fit(Data[0], Data[2])
    Predictions = clf.predict(Data[0])
    train_acc_perceptron = accuracy_score(Data[2], Predictions)
    print('The accuracy of the Random Forest classifier on the training set is %s' % train_acc_perceptron)
    Score = clf.score(Data[1], Data[3])
    print('Score from Random Forest Classifier: %s' % Score)


RanForest([train_X, test_X, train_y, test_y])
```

```python
def DecisionTree(Data):
  clf = DecisionTreeClassifier(random_state=0)
  clf.fit(Data[0], Data[2])
  Score = clf.score(Data[1], Data[3])
  print('Score from Decision Tree: %s' % Score)
DecisionTree([train_X, test_X, train_y, test_y])
```

```python
def AdaBoost(Data):
    clf = AdaBoostClassifier(n_estimators=500, learning_rate=0.5)
    # Default param - 0.37769
    clf.fit(Data[0], Data[2])
    # predict = clf.predict(test_X)
    Score = clf.score(Data[1], Data[3])
    print('Score from AdaBoost Classifier: %s' % Score)
AdaBoost([train_X, test_X, train_y, test_y])
```

```python
def GradientBoosting(Data):
    clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
    clf.fit(Data[0], Data[2])
    Score = clf.score(Data[1], Data[3])
    print('Score from Gradient Boosting Classifier: %s' % Score)
GradientBoosting([train_X, test_X, train_y, test_y])
```

```python
def Bagging(Data):
    clf = BaggingClassifier(base_estimator=SVC(), n_estimators=10, random_state=0)
    clf.fit(Data[0], Data[2])
    Score = clf.score(Data[1], Data[3])
    print('Score from Bagging Classifier: %s' % Score)
Bagging([train_X, test_X, train_y, test_y])
```
